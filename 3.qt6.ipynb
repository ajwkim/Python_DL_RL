```python
# custom signal
# upbit API 적용에 적절해 보임

import time
from PyQt6.QtCore import QTimer, QThreadPool, pyqtSlot, pyqtSignal
from PyQt6.QtWidgets import (QApplication, QMainWindow, QWidget, QVBoxLayout, 
                             QLabel, QPushButton)


class MainWindow(QMainWindow):
    custom_signal = pyqtSignal()                                    # class attr

    def __init__(self):
        super().__init__()

        self.threadpool = QThreadPool()                                 # QThreadPool()
        print('Multithreading with maximum {} threads.'.format(
            self.threadpool.maxThreadCount()))                          # .maxThreadCount()
        
        self.custom_signal.connect(self.signal_handler)             # custom_signal.connect

        self.counter = 0
        self.label = QLabel('Start')
        button = QPushButton('Danger!')
        button.pressed.connect(self.oh_no)

        layout = QVBoxLayout()
        layout.addWidget(self.label)
        layout.addWidget(button)
        w = QWidget()
        w.setLayout(layout)
        self.setCentralWidget(w)
        self.show()

        self.timer = QTimer()
        self.timer.setInterval(1000)
        self.timer.timeout.connect(self.recurring_timer)
        self.timer.start()

    def oh_no(self):
        self.threadpool.start(self.some_work)                        # threadpool.start(some_work)

    @pyqtSlot()
    def some_work(self):
        print('\nThread start')
        self.custom_signal.emit()                       # custom_signal.emit() -> main thread
        for _ in range(5):
            time.sleep(1)
        self.counter -= 10
        print('Thread complete')

    def signal_handler(self):
        print('Signal received')

    def recurring_timer(self):
        self.counter += 1
        self.label.setText(f'Counter: {self.counter}')


app = QApplication([])
window = MainWindow()
app.exec()
```

```python
# thread IO

import sys, time, random
from PyQt6.QtCore import (QObject, QRunnable, QThreadPool, pyqtSignal, pyqtSlot, QTimer)
from PyQt6.QtWidgets import (QApplication, QMainWindow, QWidget, QVBoxLayout, QLabel, QPushButton)


class WorkerSignal(QObject):        # worker thread에서 사용할 signal
    finished = pyqtSignal()
    result = pyqtSignal(dict)
    error = pyqtSignal(str)


class Worker(QRunnable):                    # QRunnable
    def __init__(self, iterations=5):
        super().__init__()
        self.signals = WorkerSignal()
        self.iterations = iterations

    @pyqtSlot()
    def run(self):
        try:
            for i in range(self.iterations):
                time.sleep(.01)
                v = 5 / (40 - i)
        except Exception as e:
            self.signals.error.emit(str(e))
        else:
            self.signals.finished.emit()
            self.signals.result.emit({'n': i, 'value': v})


class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.threadpool = QThreadPool()                             # QThreadPool
        print('Multithreadign with maximum {} threads'.format(
            self.threadpool.maxThreadCount()))
        
        self.counter = 0
        self.l = QLabel('Start')
        b = QPushButton('Danger!')
        b.pressed.connect(self.oh_no)

        layout = QVBoxLayout()
        layout.addWidget(self.l)
        layout.addWidget(b)
        w = QWidget()
        w.setLayout(layout)
        self.setCentralWidget(w)
        self.show()

        self.timer = QTimer()
        self.timer.setInterval(1000)
        self.timer.timeout.connect(self.recurring_timer)
        self.timer.start()

    def oh_no(self):
        worker = Worker(random.randrange(10, 51))
        worker.signals.result.connect(self.worker_output)
        worker.signals.finished.connect(self.worker_complete)
        worker.signals.error.connect(self.worker_error)
        self.threadpool.start(worker)                           # threadpool.start(worker)

    def worker_output(self, s):
        print('RESULT', s)

    def worker_complete(self):
        print('THREAD COMPLETE')
    
    def worker_error(self, t):
        print('ERROR:', t)

    def recurring_timer(self):
        self.counter += 1
        self.l.setText(f'Counter: {self.counter}')


app = QApplication(sys.argv)
window = MainWindow()
app.exec()
```

#### progress watcher

```python
import sys, time, random, uuid
from PyQt6.QtCore import (QObject, QRunnable, QThreadPool, pyqtSignal, pyqtSlot, QTimer)
from PyQt6.QtWidgets import (QApplication, QMainWindow, QWidget, QVBoxLayout,
                             QLabel, QPushButton, QProgressBar)


class WorkerSignals(QObject):
    progress = pyqtSignal(str, int)
    finished = pyqtSignal(str)


class Worker(QRunnable):
    def __init__(self):
        super().__init__()
        self.job_id = uuid.uuid4().hex
        self.signals = WorkerSignals()

    @pyqtSlot()
    def run(self):
        total = 1000
        delay = random.random() / 100
        for i in range(total):
            progress_pc = int(100 * (i+1) / total)
            self.signals.progress.emit(self.job_id, progress_pc)
            time.sleep(delay)
        self.signals.finished.emit(self.job_id)


class MainWindow(QMainWindow):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

        self.progress = QProgressBar()
        button = QPushButton('START IP UP')
        button.pressed.connect(self.execute)
        self.status = QLabel('0 workers')

        layout = QVBoxLayout()
        layout.addWidget(self.progress)
        layout.addWidget(button)
        layout.addWidget(self.status)
        w = QWidget()
        w.setLayout(layout)
        self.setCentralWidget(w)
        self.show()

        self.worker_progress = {}
        self.threadpool = QThreadPool()
        print('Multithreading with maximum {} threads'.format(self.threadpool.maxThreadCount()))

        self.timer = QTimer()
        self.timer.setInterval(100)
        self.timer.timeout.connect(self.refresh_progress)
        self.timer.start()

    def execute(self):
        worker = Worker()
        worker.signals.progress.connect(self.update_progress)
        worker.signals.finished.connect(self.cleanup)
        self.threadpool.start(worker)

    def cleanup(self, job_id):
        if all(v==100 for v in self.worker_progress.values()):
            self.worker_progress.clear()
            self.refresh_progress()

    def update_progress(self, job_id, progress):
        self.worker_progress[job_id] = progress
        
    def calculate_progress(self):
        if not self.worker_progress:
            return 0
        return int(sum(v for v in self.worker_progress.values()) / len(self.worker_progress))
    
    def refresh_progress(self):
        progress = self.calculate_progress()
        print(self.worker_progress)
        self.progress.setValue(progress)
        self.status.setText(f'{len(self.worker_progress)} workers')


app = QApplication(sys.argv)
window = MainWindow()
app.exec()
```

## QThread

```python
import sys, time
from PyQt6.QtCore import QThread, pyqtSignal, pyqtSlot
from PyQt6.QtWidgets import QApplication, QMainWindow, QLabel, QPushButton, QWidget, QVBoxLayout


class Thread(QThread):
    result = pyqtSignal(str)

    @pyqtSlot()
    def run(self):
        print('Thread start')
        counter = 0
        while True:
            time.sleep(1)
            self.result.emit(f'The number is {counter}')        # 별도 thread에서 signal emit
            counter += 1
        print('Thread complete')

    
class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.thread = Thread()

        label = QLabel('Output will appear here')
        button = QPushButton('Kill thread')
        button.pressed.connect(self.thread.terminate)               # .terminate()
        container = QWidget()
        layout = QVBoxLayout()
        layout.addWidget(label)
        layout.addWidget(button)
        container.setLayout(layout)
        self.setCentralWidget(container)
        self.show()

        self.thread.start()                                         # .start()
        self.thread.result.connect(label.setText)                   # signal connect
        self.thread.finished.connect(self.thread_has_finished)      # finished signal

    def thread_has_finished(self):
        print('Thread has finished')
        print(self.thread, self.thread.isRunning(), self.thread.isFinished())   # .isRunning(), .isFinished()


app = QApplication(sys.argv)
window = MainWindow()
app.exec()
```

```python
# 데이터 전송

import sys, time
from PyQt6.QtCore import QThread, pyqtSignal, pyqtSlot
from PyQt6.QtWidgets import QApplication, QMainWindow, QWidget, QVBoxLayout, QLabel, QPushButton, QSpinBox


class Thread(QThread):
    
    result = pyqtSignal(str)

    def __init__(self, initial_counter=0):
        super().__init__()
        self.counter = initial_counter

    @pyqtSlot()
    def run(self):
        print('Thread start')
        self.is_running = True
        self.waiting_for_data = True
        while True:
            while self.waiting_for_data:
                if not self.is_running:
                    return
                time.sleep(.1)
            self.counter += self.input_add
            self.counter *= self.input_mul
            self.result.emit(f'The cumulative total is {self.counter}')
            self.waiting_for_data = True

    def send_data(self, add, mul):
        self.input_add = add
        self.input_mul = mul
        self.waiting_for_data = False

    def stop(self):
        self.is_running = False


class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.thread = Thread(500)                                      # Thread()
        self.thread.start()                                         # .start()

        self.add_input = QSpinBox()        
        self.mul_input = QSpinBox()
        button_input = QPushButton('Submit number')
        label = QLabel('Output will appear here')
        button_stop = QPushButton('Shutdown thread')
        button_stop.pressed.connect(self.thread.stop)                    # .사용자정의중단함수

        button_input.pressed.connect(self.submit_data)
        self.thread.result.connect(label.setText)                   # signal connect
        self.thread.finished.connect(self.thread_has_finished)      # finished signal

        container = QWidget()
        layout = QVBoxLayout()
        layout.addWidget(self.add_input)
        layout.addWidget(self.mul_input)
        layout.addWidget(button_input)
        layout.addWidget(label)
        layout.addWidget(button_stop)
        container.setLayout(layout)
        self.setCentralWidget(container)
        self.show()

    def submit_data(self):
        self.thread.send_data(self.add_input.value(), self.mul_input.value())

    def thread_has_finished(self):
        print('Thread has finished')
        print(self.thread, self.thread.isRunning(), self.thread.isFinished())   # .isRunning(), .isFinished()


app = QApplication(sys.argv)
window = MainWindow()
app.exec()
```
